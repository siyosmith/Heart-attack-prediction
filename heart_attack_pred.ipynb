```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score
```

```python
df=pd.read_csv("/content/drive/MyDrive/heart_attack_prediction_dataset.csv")
df
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
columns_to_drop = ["Country", "Continent", "Hemisphere","Patient ID"]
existing_columns_to_drop = [col for col in columns_to_drop if col in df.columns]
df = df.drop(existing_columns_to_drop, axis=1)
```

```python
df
```

```python
df['Sex']=df['Sex'].map({'Male':1,'Female':0})
df
```

```python
df[['Systolic', 'Diastolic']] = df['Blood Pressure'].str.split('/', expand=True)
```

```python
df.drop('Blood Pressure', axis=1, inplace=True)
```

```python
df['Diet']=df['Diet'].map({'Average':1,'Unhealthy':0,'Healthy':2})
```

```python
df['Systolic'].astype(int)
```

```python
df['Diastolic'].astype(int)
```

```python
y = df["Heart Attack Risk"]
X = df.drop('Heart Attack Risk',axis=1)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)
```

```python
scaler = StandardScaler()
scaler.fit(X_train)
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

```python
from sklearn.model_selection import GridSearchCV
parameter={'penalty':['l1','l2','elasticnet'],'C':[1,2,3,4,5,6,10,20,30,40,50],'max_iter':[100,200,300]}
```

```python
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression()
```

```python
classsifier_regressor = GridSearchCV(classifier,param_grid=parameter,scoring='accuracy',cv=5)
```

```python
classsifier_regressor.fit(X_train,y_train)
```

```python
  print(classsifier_regressor.best_params_)
```

```python
print(classsifier_regressor.best_score_)
```

```python
y_pred=classsifier_regressor.predict(X_test)
```

```python
score=accuracy_score(y_pred,y_test)
score
```
